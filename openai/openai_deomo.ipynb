## **ChatCompletion API**

```python
prompt = "Hello, how are you?"
```

```python
# Define the messages
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello!"}
]

# Make the API request using the new method
response = openai.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=messages,
    max_tokens=100
)

# Print the response
print(response)
```

```python
print(response.choices[0].message.content)
```

```python
#chat comptetion api

prompt1 = "Hello, how are you?"
prompt2 = "I am 26 years old and I am a ML engineer"
prompt3 = "Tell about me"

response = openai.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt1},
        {"role": "user", "content": prompt2},
        {"role": "user", "content": prompt3}
    ],
    max_tokens=100
)
print(response.choices[0].message.content)
```
